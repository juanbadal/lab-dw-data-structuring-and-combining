# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_extended = pd.read_csv(url)
df_extended
```

```python
# Removing 'unnamed:_0' column

df_extended = df_extended.drop('unnamed:_0', axis=1)
df_extended
```

```python
# Checking for nulls

df_extended.isna().sum()
```

```python
# Checking for dupes

print('Before:', df_extended.duplicated().sum())

# There are 443 fully duplicated rows, let's remove them

df_extended = df_extended.drop_duplicates().reset_index(drop=True)

print('After:', df_extended.duplicated().sum())
```

```python
# Checking for dupes in the 'customer' column, which appears to be a unique identifier

print(df_extended['customer'].duplicated().sum())

# Let's see some examples of the duplicated 'customer' column

display(df_extended[df_extended['customer'].duplicated() == True])

display(df_extended[df_extended['customer'].isin(['AV76907', 'FV22585', 'AU23837'])])
```

```python
# Checking if every set of duplicates from 'customer' column has the same value in 'customer_lifetime_value'

result = df_extended.groupby('customer')['customer_lifetime_value'].nunique().eq(1).all()

print(result)
```

```python
# Removing dupes in 'customer' column, keeping the last occurrence

df_extended = df_extended.drop_duplicates(subset=['customer'], keep='last').reset_index(drop=True)

# Checking for dupes in 'customer' column

df_extended['customer'].duplicated().sum()
```

```python
# Checking column data types

df_extended.dtypes
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Creating pivot table
pivot = df_extended.pivot_table(index='sales_channel', values='customer_lifetime_value', aggfunc='sum').reset_index()

# Rounding to 2 decimals
pivot['customer_lifetime_value'] = pivot['customer_lifetime_value'].round(2)

# Adding column to pivot table to show the weight of each channel
pivot['%_of_total'] = pivot['customer_lifetime_value'] / pivot['customer_lifetime_value'].sum()

pivot
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Creating pivot table
grouped = df_extended.groupby(['gender', 'education'])['customer_lifetime_value'].agg('mean')

grouped = pd.DataFrame(grouped)

grouped
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Creating wide format table with groupby

grouped_complaints = df_extended.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum()
grouped_complaints = pd.DataFrame(grouped_complaints).reset_index()
grouped_complaints
```

```python
# Melting into long format

grouped_complaints_melt = pd.melt(grouped_complaints, id_vars=['policy_type', 'month'], value_name='sum_of_open_complaints')
grouped_complaints_melt
```
